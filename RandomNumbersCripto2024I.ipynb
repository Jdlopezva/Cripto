```python
import random
import numpy as np
import matplotlib.pyplot as plt

def generar_numeros_aleatorios():
    cantidad = int(input("Ingrese la cantidad de números aleatorios a generar: "))
    semilla = int(input("Ingrese el número semilla: "))
    permitir_repeticion = input("¿Permitir repeticiones? (SI/NO): ").upper() == "SI"

    random.seed(semilla)

    # Generar primera lista de números
    numeros_generados = set() if not permitir_repeticion else []
    while len(numeros_generados) < cantidad:
        numero = random.randint(1, 100)  # Generar número aleatorio entre 1 y 100 (puedes ajustar el rango según necesites)
        if permitir_repeticion:
            numeros_generados.append(numero)
        else:
            numeros_generados.add(numero)

    print("Números generados:", numeros_generados)

    # Crear gráfico de dispersión para la primera lista de números
    plt.figure()
    plt.subplot(3, 1, 1)
    plt.scatter(range(1, cantidad+1), numeros_generados)
    plt.title("Gráfico de dispersión de números generados")
    plt.xlabel("Índice")
    plt.ylabel("Número generado")

    # Generar segunda lista de números que se asemejen a una distribución normal
    numeros_distribucion_normal = np.random.normal(0, 1, cantidad)
    numeros_distribucion_normal_ordenados = sorted(numeros_distribucion_normal)

    # Crear gráfico de dispersión para la segunda lista de números
    plt.subplot(3, 1, 2)
    plt.scatter(range(1, cantidad+1), numeros_distribucion_normal_ordenados)
    plt.title("Gráfico de dispersión de números con distribución normal")
    plt.xlabel("Índice")
    plt.ylabel("Número generado")

    # Generar tercera lista de números que se asemejen a una distribución de Poisson
    lambd = 5  # Parámetro lambda para la distribución de Poisson
    numeros_distribucion_poisson = np.random.poisson(lambd, cantidad)

    # Crear gráfico de dispersión para la tercera lista de números
    plt.subplot(3, 1, 3)
    plt.scatter(range(1, cantidad+1), numeros_distribucion_poisson)
    plt.title("Gráfico de dispersión de números con distribución de Poisson")
    plt.xlabel("Índice")
    plt.ylabel("Número generado")

    plt.tight_layout()
    plt.show()

generar_numeros_aleatorios()
```
